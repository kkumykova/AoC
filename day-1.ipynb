```python
# split the string into an array using the newline character as the delimiter
# https://www.w3schools.com/python/ref_string_split.asp

split_input = raw_input.split("\n")
split_input
```

```python
# split_input is an array of strings which needs to be converted into integer values
# https://realpython.com/list-comprehension-python/
expenses = [int(string) for string in split_input]
expenses
```

```python
# https://docs.python.org/3/library/itertools.html
# combinations() method from Python’s itertools module
from itertools import *
all_expense_pairs = list(combinations(expenses, 2))
len(all_expense_pairs)
```

```python
# Any time you’re in a situation where you need to find values in an array that match some criteria,
# you should think about applying a filter() function.

def sums_to_2020(values):
    return sum(values) == 2020

result = list(filter(sums_to_2020, all_expense_pairs))
result
# The resulting list had one tuple:
```

```python
all_expense_triplets = list(combinations(expenses, 3))
result2 = list(filter(sums_to_2020, all_expense_triplets))
result2
```
